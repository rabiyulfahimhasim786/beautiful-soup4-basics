```python
import requests, bs4
from bs4 import BeautifulSoup
from csv import writer
```

```python
import requests
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:96.0) Gecko/20100101 Firefox/96.0'}
HOST = 'https://www.zocdoc.com'
PAGE = 'search?address=&insurance_carrier=-1&insurance_plan=-1&searchType=specialty&reason_visit=386&search_query=Gastroenterologist&dr_specialty=106&ppsSelectionId=e10fceed-aedf-4b0e-94a9-5abb3397e1be'
with requests.Session() as session:
    (r := session.get(HOST, headers=headers)).raise_for_status()
    (r := session.get(f'{HOST}/{PAGE}', headers=headers)).raise_for_status()
    # process content from here
print(r.text)
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
soup
print(soup.prettify())
```

```python
#https://www.zocdoc.com/search?address=&insurance_carrier=-1&insurance_plan=-1&searchType=specialty&reason_visit=386&search_query=Gastroenterologist&dr_specialty=106&ppsSelectionId=e10fceed-aedf-4b0e-94a9-5abb3397e1be
```

```python
b = soup.find_all('span', attrs={'data-test': 'doctor-card-info-name-full'})

doctor = []

for a in b:
 doctor.append(a.text.strip())
print(doctor)
```

```python
len(doctor)
```

```python
c = soup.find_all('div', attrs={'data-test': 'doctor-card-info-specialty'})

specialty = []

for d in c:
 specialty.append(d.text.strip())
print(specialty)
```

```python
len(specialty)
```

```python
e = soup.find_all('div', attrs={'data-test': 'doctor-card-info-location'})

location = []

for f in e:
 location.append(f.text.strip())
print(location)
```

```python
len(location)
```

```python
g = soup.find_all('div', attrs={'data-test': 'doctor-card-info-provider-qualities'})

details = []

for h in g:
 details.append(h.text.strip())
print(details)
```

```python
len(details)
```
