```python
import requests, bs4
from bs4 import BeautifulSoup
from csv import writer
```

```python
url = "https://posterswholesale.americommerce.com/store/c/29-Sports.aspx"
page = requests.get(url)
page
page.status_code
```

```python
page.content
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
soup
print(soup.prettify())
```

```python
movies_html = soup.find_all('a', class_='h5 color-inherit')

movie_titles = []

for title in movies_html:
 movie_titles.append(title.text.strip())
print(movie_titles)
```

```python
stocks_html = soup.find_all('span', class_='CategoryProductAvailability')

stock_titles = []

for title in stocks_html:
 stock_titles.append(title.text.strip())
print(stock_titles)
```

```python
items_html = soup.find_all('span', class_='CategoryProductItemNr')

item_titles = []

for title in items_html:
 item_titles.append(title.text.strip())
print(item_titles)
```

```python
price_html = soup.find_all('span', class_='CategoryProductRetailPrice')

price_titles = []

for title in price_html:
 price_titles.append(title.text.strip())
print(price_titles)
```
