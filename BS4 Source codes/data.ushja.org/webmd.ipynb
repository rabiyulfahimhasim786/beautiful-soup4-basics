```python
from bs4 import BeautifulSoup as bs
import requests 
import csv
import pandas as pd

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:98.0) Gecko/20100101 Firefox/98.0'
}
link = 'https://doctor.webmd.com/doctor/priscilla-garcia-79cc78e2-d27b-45f9-a767-5ae0c34af2d2-overview'
      
page = requests.get(link, headers = headers)

# getting content from link
page.content

soup = bs(page.content, 'html.parser')
#it gives us the visual representation of data

a = soup.find_all('h1', class_="provider-full-name loc-co-fname")
b = soup.find_all('div', class_="prov-specialties-wrap")
c = soup.find_all('div', class_="prov-location-name loc-co-tplcnm")
d = soup.find_all('p', class_="prov-address-text loc-co-tplcadd")
#e = soup.find_all('button', class_="webmd-button phone-btn loc-co-tpphn webmd-button--ghost webmd-button--medium")
e = soup.find_all('div', class_="btn-holder")
f = soup.find_all('img', class_="loc-co-provim")

Doctor_name = []
Specialty  = []
Hospitals = []
Address = []
Phone_Number = []
Image_link = []

for title in a:
  Doctor_name.append(title.text.strip())
for title in b:
 Specialty.append(title.text.strip())
for title in c:
  Hospitals.append(title.text.strip())
for title in d:
 Address.append(title.text.strip())
for title in e:
 Phone_Number.append(title.text.strip())
for title in f:
  Image_link.append(title['src'])

print(Doctor_name)
print(Specialty)
print(Hospitals)
print(Address)
print(Phone_Number)
print(Image_link)

import pandas as pd
df = pd.DataFrame({'A':['1','2'],
                   'B': ['2','3'],
                   'C': ['3','4'],
                   'D': ['4','5'],
                   'E': ['5','6'],
                   'F': ['6','7']})
df

import pandas as pd
df = pd.DataFrame({'Doctor_name':Doctor_name,
                   'Specialty':Specialty,
                   #'Hospitals':Hospitals,
                   'Address':Address,
                   'Phone_Number':Phone_Number})
                   #'Image_link':Image_link})
df
```
