```python
import requests, bs4
from bs4 import BeautifulSoup
from csv import writer
```

```python
url = "https://www.desss.com/internet-of-things-IoT-consulting-advisory.html"
page = requests.get(url)
page
page.status_code
```

```python
page.content
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
soup
print(soup.prettify())
```

```python
a = soup.find_all('p', class_='f-h3-head')

Product_Titles = []

for c in a:
 Product_Titles.append(c.text.strip())
print(Product_Titles)
```

```python
a_html = soup.find_all('a', class_='')

base_titles = []

for title in a_html:
 base_titles.append(title.text.strip())
print(base_titles)
```

```python
a_html = soup.find_all('div', class_='center-align black-text heading')

iot_titles = []

for title in a_html:
 iot_titles.append(title.text.strip())
print(iot_titles)
```

```python
a_html = soup.find_all('div', class_='introducation-head justify-align black-light-text')

define_titles = []

for title in a_html:
 define_titles.append(title.text.strip())
print(define_titles)
```

```python
a_html = soup.find_all('div', class_='footer-copyrights center')

b_titles = []

for title in a_html:
 b_titles.append(title.text.strip())
print(b_titles)
```

```python
a_html = soup.find_all('div', class_='footer-all-details')

business_titles = []

for title in a_html:
 business_titles.append(title.text.strip())
print(business_titles)
```

```python
a_html = soup.find_all('div', class_='col -lg-12')

contact_titles = []

for title in a_html:
 contact_titles.append(title.text.strip())
print(contact_titles)
```

```python
import pandas as pd
import xlsxwriter


# dataframe Name and Age columns
df = pd.DataFrame({'Product_Titles': Product_Titles,
                   #'Base_titles': base_titles,
                   #'iot_titles': iot_titles,
                   #'business_titles': business_titles,
                   #'Copyright Title': b_titles,
                   #'Define_titles': define_titles,
                   })


# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter('test5.xlsx', engine='xlsxwriter')

# Convert the dataframe to an XlsxWriter Excel object.
df.to_excel(writer, sheet_name='Sheet1', index=False)


# Close the Pandas Excel writer and output the Excel file.
writer.save()
```
