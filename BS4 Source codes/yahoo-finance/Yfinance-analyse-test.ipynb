```python
import yfinance as yahooFinance
 
 
GetFacebookInformation = yahooFinance.Ticker("FB")

#previousClose

print("Previous Close : ", GetFacebookInformation.info['previousClose'])

#Open
print("Open : ", GetFacebookInformation.info['open'])

#bidSize
print("bid : ", GetFacebookInformation.info['bid'])
print("bidSize : ", GetFacebookInformation.info['bidSize'])

print("ask : ", GetFacebookInformation.info['ask'])
print("askSize : ", GetFacebookInformation.info['askSize'])
 
#Open
print("days range high : ", GetFacebookInformation.info['regularMarketDayHigh'])

#Open
print("days range Low : ", GetFacebookInformation.info['regularMarketDayLow'])
#Open
print("fiftyTwoWeekLow : ", GetFacebookInformation.info['fiftyTwoWeekLow'])
print("fiftyTwoWeekHigh : ", GetFacebookInformation.info['fiftyTwoWeekHigh'])
# display Company Sector
print("regularMarketVolume : ", GetFacebookInformation.info['regularMarketVolume'])

 
# display Company Beta
print("volume : ", GetFacebookInformation.info['volume'])

# display Price Earnings Ratio
print("averageVolume : ", GetFacebookInformation.info['averageVolume'])

# display Price Earnings Ratio
print("marketCap : ", GetFacebookInformation.info['marketCap'])

print("beta (5Y Monthly): ", GetFacebookInformation.info['beta'])

print("trailingPE : ", GetFacebookInformation.info['trailingPE'])

print("trailingEps : ", GetFacebookInformation.info['trailingEps'])

#trailingPE
#print("targetMeanPrice : ", GetFacebookInformation.info['targetMeanPrice'])

print("fiveYearAvgDividendYield : ", GetFacebookInformation.info['fiveYearAvgDividendYield'])
#fiveYearAvgDividendYield
print("exDividendDate : ", GetFacebookInformation.info['exDividendDate'])

print("1y Target Est : ", GetFacebookInformation.info['targetMeanPrice'])
```

```python
import yfinance as yahooFinance
 
 
GetFacebookInformation = yahooFinance.Ticker("FB")
 
# Let us  get historical stock prices for Facebook
# covering the past few years.
# max->maximum number of daily prices available
# for Facebook.
# Valid options are 1d, 5d, 1mo, 3mo, 6mo, 1y, 2y,
# 5y, 10y and ytd.
print(GetFacebookInformation.history(period="max"))
```

```python
import yfinance as yahooFinance
 
# in order to specify start date and
# end date we need datetime package
import datetime
 
# startDate , as per our convenience we can modify
startDate = datetime.datetime(2019, 5, 31)
 
# endDate , as per our convenience we can modify
endDate = datetime.datetime(2021, 1, 30)
GetFacebookInformation = yahooFinance.Ticker("FB")
 
# pass the parameters as the taken dates for start and end
print(GetFacebookInformation.history(start=startDate,
                                     end=endDate))
```
