```python
import requests
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:96.0) Gecko/20100101 Firefox/96.0'}
HOST = 'https://finance.yahoo.com/gainers?count=100&offset=100'
#PAGE = 'primaryschools'
with requests.Session() as session:
    (r := session.get(HOST, headers=headers)).raise_for_status()
   # (r := session.get(f'{HOST}/{PAGE}', headers=headers)).raise_for_status()
    (r := session.get(f'{HOST}', headers=headers)).raise_for_status()
    # process content from here
print(r.text)
```

```python
soup = BeautifulSoup(r.text, 'lxml')
soup
print(soup.prettify())
```

```python
movies_html = soup.find_all('td', attrs={'aria-label': 'Name'})

doctor_n = []

for title in movies_html:
 doctor_n.append(title.text.strip())
print(doctor_n)
```
