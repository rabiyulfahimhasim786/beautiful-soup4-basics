```python
import requests
```

```python
BASE_URL = 'https://www.healthgrades.com/api3/usearch'
```

```python
params = {
    'userLocalTime':'22:37',
    'what':'Family+Medicine',
    'where':'Houston, TX',
    #'pt':'40.720901, -74.008904',
    'sort.provider':'bestmatch',
    'category':'provider',
    'sessionId':'S4e4d',
    'requestId':'a948ae51ff1009d6',
    'pageSize.provider':'239',#239
    'pageNum':'11',#11
    'isFirstRequest':'true',
    'debug':'false',
    'isAtlas':'true',
    'action':'refresh',
    #'neCorner':'40.744526282819244,-73.99060337556104',
    #'swCorner':'40.69723118200452,-74.02724113269275'
    }
```

```python
r = requests.get(BASE_URL, params=params)
print(r.text)
```

```python
r.raise_for_status()
```

```python
dentists = r.json()['search']['searchResults']['provider']['results']
```

```python
for dentist in dentists:
  print(dentist['displayName'])
  print(dentist['specialistDesc'])
  print(dentist['addresses'][0]['address1'])
  print(dentist['addresses'][0]['officeCode'])
  print(dentist['addresses'][0]['officeGuid'])
  print(dentist['addresses'][0]['location']['cityName'])
  print(dentist['addresses'][0]['location']['cityAndState'])
  print(dentist['addresses'][0]['location']['cityStateZipBestMatch'])
  # print(dentist['addresses'][0]['location']['cityName']['cityAndState'])
  # print(dentist['addresses'][0]['location']['cityName']['cityStateZipBestMatch'])
```

```python
import pandas as pd
data_all=[]
for dentist in dentists:
  doctor_name = dentist['displayName']
  specialty = "Family Practitioner" #print(dentist['specialistDesc'])
  address =dentist['addresses'][0]['address1']
  officeCode = dentist['addresses'][0]['officeCode']
  officeGuid = dentist['addresses'][0]['officeGuid']
  cityName =dentist['addresses'][0]['location']['cityName']
  cityAndState = dentist['addresses'][0]['location']['cityAndState']
  cityStateZipBestMatch = dentist['addresses'][0]['location']['cityStateZipBestMatch']
  data_all.append({'Doctor Name':doctor_name,'Price':specialty,'Address': address, 'OfficeCode': officeCode, 'OfficeGuid': officeGuid, 'CityName': cityName, 'CityAndState': cityAndState, 'CityStateZipBestMatch':cityStateZipBestMatch})
df = pd.DataFrame(data_all)
df.to_csv('products11.csv',encoding='utf-8')
  # print(dentist['addresses'][0]['location']['cityName']['cityAndState'])
  # print(dentist['addresses'][0]['location']['cityName']['cityStateZipBestMatch'])
```
